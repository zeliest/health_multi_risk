```python
import xarray as xr
file = "/Users/szelie/Library/CloudStorage/OneDrive-ETHZurich/data/health_multi_risk_data/PMfire_glob_2003to2021.nc4"
fire = xr.open_dataset(file)
```

### fire

```python
fire = fire.rename({'lat':'latitude', 'lon':'longitude'})
```

```python
fire_monthly = fire.resample(time='M').max()
```

```python
import pandas as pd
import numpy as np
from climada.hazard import Hazard
# Rename columns

# Extract years from the `time` dimension
years = pd.to_datetime(fire['time'].values).year
unique_years = np.unique(years)

# Initialize an empty list to store the hazards
yearly_hazards = {}

# Loop over each unique year
for year in unique_years:
    # Extract data for that year
    mask = (pd.to_datetime(fire_monthly['time'].values).year == year)
    yearly_fire_data = fire_monthly.isel(time=mask)

    # Convert it into a hazard
    yearly_hazard = Hazard.from_xarray_raster(yearly_fire_data, hazard_type='pm25', intensity_unit='kg/m3', intensity='cnc_PM_FRP_m_17')
    yearly_hazard.intensity.data[yearly_hazard.intensity.data<1e-9] = 0
    yearly_hazard.intensity.eliminate_zeros()
    # Append the hazard to the list
    yearly_hazards[year] = yearly_hazard

## Store the hazards in the hazards dictionary
#hazards['pm25_fire'] = Hazard.concat(yearly_hazards)
```

```python
from pathlib import Path
from climada.entity import Exposures
import numpy as np
exposures_dict = {}
path_exposures = Path("/Users/szelie/OneDrive - ETH Zurich/data/health_multi_risk_data/exposures")


for age in ['0_1']:#,'65_70_75_80','all']:
    exposures_dict[age] = {}
    for year in np.arange(2003, 2023):
        year = str(year)
        exposures_dict[age][year] = Exposures.from_hdf5(path_exposures / f"{age}_era5_025_compatible_{year}.hdf5")
        exposures_dict[age][year].gdf["impf_pm25"] = 1 
        exposures_dict[age][year].gdf['longitude'] = exposures_dict[age][year].gdf['longitude'].apply(lambda x: x - 360 if x > 180 else x)
        exposures_dict[age][year].gdf['value'][exposures_dict[age][year].gdf['value']<0] = 0
```

```python
exposures_dict['65_70_75_80']['2000'].gdf
```

```python
from climada.entity.impact_funcs import ImpactFunc, ImpactFuncSet

impf_set = ImpactFuncSet()

impf25 = ImpactFunc()
impf25.id = 1
impf25.haz_type = 'pm25'
impf25.name = 'Step function pm2.5'
impf25.intensity_unit = ''
impf25.intensity = np.array([0, 1.4e-8, 1.5e-8, 1])
impf25.mdd = np.array([0, 0, 1, 1])
impf25.paa = np.ones(impf25.intensity.shape)
impf25.check()
impf_set.append(impf25)
```

```python
yearly_hazards[year].intensity
```

```python
from climada.engine import Impact, ImpactCalc

impact_dict = {age: {year:ImpactCalc(exposures_dict[age][str(year)], impf_set, yearly_hazards[year]).impact(save_mat=True) for year in yearly_hazards} for age in exposures_dict}


impact_dict_no_change_vuln_pop = {age: {year:ImpactCalc(exposures_dict[age]['2003'], impf_set, yearly_hazards[year]).impact(save_mat=True) for year in yearly_hazards} for age in exposures_dict}

for age in impact_dict:
    for year in impact_dict[age]:
        impact_dict[age][year].tot_value=0
        impact_dict_no_change_vuln_pop[age][year].tot_value=0

```

```python
impact_concat = {}
impact_concat_no_change_vuln_pop = {}
for age in impact_dict:
    impact_concat[age] = Impact.concat(list(impact_dict[age].values()),reset_event_ids=True)
    impact_concat_no_change_vuln_pop[age] = Impact.concat(list(impact_dict_no_change_vuln_pop[age].values()),reset_event_ids=True)
```

```python
impact_concat['0'].at_event
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
date = [datetime.fromordinal(date) for date in impact_concat[age].date]

# Sample data for demonstration
# (Replace with your actual data)
# date = ...
# impact_concat = ...
# impact_concat_no_change_vuln_pop = ...

fig, ax = plt.subplots(figsize=(14,6))
ax.scatter(date, impact_concat[age].at_event, alpha=0.8, label="changes in climatee& population")
ax.scatter(date, impact_concat_no_change_vuln_pop[age].at_event, alpha=0.8, label="changes in climate")

# Identify the indices of the top 8 maximum values for impact_concat
sorted_indices = impact_concat[age].at_event.argsort()
top_8_indices = sorted_indices[-8:]  # indices of the top 8 maximum values

# Annotate the top 8 maximum values for impact_concat
for idx in top_8_indices:
    ax.annotate(date[idx], (date[idx], impact_concat[age].at_event[idx]), fontsize=9, ha='right')

ax.set_title("Change in vulnerable population exposed (only climate)")
ax.legend()
plt.show()
```
