```python
import itertools
from climada.engine import Impact
combinations = []

hazards = ['RF_2y','EP','DR','HW','FI','TC']

for r in range(1, len(hazards) + 1):
    combinations.extend(list(itertools.combinations(hazards, r)))

combinations = ["-".join(combi) for combi in combinations]
```

```python
from pathlib import Path
IMPACT_DIR = Path("/Users/szelie/OneDrive - ETH Zurich/data/health_multi_risk_data/impacts/multi_haz")


imp_combined = {}
for age in ['all']:
    imp_combined[age] = {}
    for combi in combinations:
        imp_combined[age][combi] = Impact.from_csv(IMPACT_DIR / f"csv/{combi}_{age}_2003_2022.csv")
        imp_combined[age][combi].imp_mat = Impact.read_sparse_csr(IMPACT_DIR / f"npz/{combi}_{age}_2003_2022.npz")
        imp_combined[age][combi]
```

```python
from pathlib import Path
combinations_ec = []
hazards = ['RF_2y_ec','EP_ec','DR_ec','HW_ec','FI_ec','TC_ec']

for r in range(1, len(hazards) + 1):
    combinations_ec.extend(list(itertools.combinations(hazards, r)))

combinations_ec = ["-".join(combi) for combi in combinations_ec]

imp_combined_ec = {}
for age in ['all']:
    imp_combined_ec[age] = {}
    for combi in combinations_ec:
        imp_combined_ec[age][combi] = Impact.from_csv(IMPACT_DIR / f"csv/{combi}_{age}_2003_2022.csv")
        imp_combined_ec[age][combi].imp_mat = Impact.read_sparse_csr(IMPACT_DIR / f"npz/{combi}_{age}_2003_2022.npz")
        imp_combined_ec[age][combi]
```

```python
import pandas as pd
import datetime 
dfs = []

for combi in imp_combined[age]:
    combi_ec = "-".join(["".join([c, "_ec"]) for c in combi.split("-")])
    df = pd.DataFrame({
                        'date': imp_combined[age][combi].date.astype(int),
                        'at_event_population_growth': imp_combined[age][combi].at_event,
                        'at_event_constant_population': imp_combined_ec[age][combi_ec].at_event,
                        'combination': combi,
                        'age_group': age
                    })
    dfs.append(df)
df_concat = pd.concat(dfs)
df_concat['date'] = [datetime.datetime.fromordinal(int(date)) for date in df_concat.date]

df_concat.to_csv(IMPACT_DIR / "monthly_exposure_global_all_combinations_2003_2021.csv")
```

```python
import geopandas as gpd
df_concat = pd.read_csv(IMPACT_DIR / "monthly_exposure_global_all_combinations_2003_2021.csv", index_col=0)
df_concat['date'] = pd.to_datetime(df_concat['date'])

# Load the world map to use as a land mask
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Define colors for each combination
colors = {
    "FI": "orange", 
    "HW": "red", 
    "DR": "black", 
    "TC": "purple", 
    "RF_2y": "blue", 
    "EP": "lightblue",
    "HW-FI": "green", 
    "DR-HW-FI": "brown"
}

# Filter combinations based on criteria
filtered_combis = ['RF_2y', 'EP', 'DR', 'HW', 'FI', 'TC', 'RF_2y-EP', 'RF_2y-DR', 'RF_2y-HW', 'RF_2y-FI', 'RF_2y-TC', 'EP-DR', 'EP-HW', 'EP-FI', 'EP-TC', 'DR-HW', 'DR-FI', 'HW-FI', 'HW-TC', 'EP-DR-HW', 'EP-DR-FI', 'EP-HW-FI', 'EP-HW-TC', 'DR-HW-FI']

# Create a figure with 8 rows and 3 columns
fig, axes = plt.subplots(6, 4, figsize=(20, 20), gridspec_kw={'hspace': 0.15})  # Adjust hspace here to reduce row spacing
axes = axes.flatten()

# Iterate over the valid combinations and plot
for idx, combi in enumerate(filtered_combis):
    if idx >= len(axes):
        break
    ax = axes[idx]
    df_combi = df_concat[df_concat.combination == combi]

    date = df_combi.date
    mask_growth = df_combi.at_event_population_growth > 0
    mask_constant = df_combi.at_event_constant_population > 0
    ax.scatter(df_combi.date[mask_growth], df_combi.at_event_population_growth[mask_growth] / 1e6, alpha=0.8, label="With population growth", color=colors.get(combi.replace('_2y', ''), 'black'), marker='*')
    ax.scatter(date[mask_constant], df_combi.at_event_constant_population[mask_constant] / 1e6, alpha=0.8, label="Constant population (2003 levels)", color="grey", marker='x')

    # Identify the indices of the top max values
    sorted_indices = df_combi.at_event_population_growth.argsort()
    top_indices = sorted_indices[-4:]  # Indices of the top values, change to whatever values to have dates shown on plot

    # Annotate the top maximum values
    for top_idx in top_indices:
        ax.annotate(df_combi.date.iloc[top_idx].date(), (df_combi.date.iloc[top_idx], df_combi.at_event_population_growth.iloc[top_idx] / 1e6), fontsize=9, ha='right')

    ax.set_title(combi.replace('_2y', ''))
    if idx // 3 == 7:  # Only set the xlabel for the bottom row
        ax.set_xlabel("Date")
    else:
        ax.set_xticklabels([])
    if idx % 4 == 0:  # Only set the ylabel for the first column
        ax.set_ylabel("Pop. exposed (millions)")
    else:
        ax.set_yticklabels([])
    ax.tick_params(axis='x', rotation=45)

# Create a single legend for all subplots
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=12, title='Legend', title_fontsize=14, bbox_to_anchor=(0.5, 0.97), ncol=4)

# Hide any unused subplots
for ax in axes[len(filtered_combis):]:
    ax.set_visible(False)

plt.tight_layout()
plt.savefig(FIGURES / "trends.png", dpi=300, bbox_inches='tight')
plt.show()
```
